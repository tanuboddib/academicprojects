```python
import requests
import pandas as pd
import json
import os
import time

from collections import defaultdict

# Set the endpoint URL and parameters
url = 'https://www.alphavantage.co/query'

# Get the interested NYSE symbols.
interested_symbols_data = pd.read_excel("/Users/reddybr/Desktop/NYSE_interested_symbols.xlsx")
symbols = interested_symbols_data.Symbol.tolist()

for sym in symbols: 
    params = {
        'function': 'TIME_SERIES_DAILY_ADJUSTED',
        'symbol': sym,
        'outputsize':'full',
        'slice': 'year5month1',
        'datatype': 'csv',
        'apikey': '7MGTVFP1UE3SGFKX'
    }

    # Send a request to Alpha Vantage API and retrieve the response
    response = requests.get(url, params=params)
    string = response.content.decode('utf-8')

    rows = string.split("\n");

    monthly_data = defaultdict(defaultdict)
    data_header = rows[0]
    print(sym +" has " + str(len(rows)))
    #print(rows)
    for i in range(1, len(rows)):
        date = rows[i].split(",")[0].split("-")
        if len(date) < 3:
            continue
        year = date[0]
        month = date[1]
    
        if year not in monthly_data.keys():
            monthly_data[year] = defaultdict(list)
        else:
            if month not in monthly_data[year].keys():
                monthly_data[year][month] = list()

        attrs = rows[i].split(",")
        attrs[-1] = attrs[-1].replace('\r', '')
        monthly_data[year][month].append(attrs)

    for year in monthly_data.keys():
        directory = "/Users/reddybr/Downloads/DependableAI/Project/Data/" + year
        for month in monthly_data[year].keys():
            month_directory = directory + "/" + month
            if not os.path.isdir(directory):
                os.system("mkdir -p " + month_directory)
            monthly_data[year][month] = monthly_data[year][month][::-1]
            file_data = pd.DataFrame(monthly_data[year][month])
            file_data.to_csv(month_directory + "/" + sym + ".csv", header = data_header)
            print("Completed dumping data for: " + sym + " and date: " + year + "-" + month)

    time.sleep(20)
```
